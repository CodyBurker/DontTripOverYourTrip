```python
philly_users = pd.read_feather('user_philly.feather')
philly_reviews = pd.read_feather('review_philly.feather')
philly_bus = pd.read_feather('business_philly.feather')
```

```python
user_columns = ['user_id', 'bucketed_gender_scores', 'bucketed_philly_share_of_reviews', 'bucketed_average_stars_given', 'bucketed_philly_reviews', 'positive_reviews_over_total_reviews']
slim_philly_users = philly_users[user_columns]
slim_philly_users['bucketed_positive_reviews_over_total_reviews'] = np.array(np.round(slim_philly_users['positive_reviews_over_total_reviews'], 1))
slim_philly_users['bucketed_positive_reviews_over_total_reviews'] = slim_philly_users['bucketed_positive_reviews_over_total_reviews'].astype(str)
slim_philly_users = slim_philly_users.drop(labels=['positive_reviews_over_total_reviews'], axis=1)
slim_philly_users
```

```python
slim_philly_users_columns = ['bucketed_gender_scores', 'bucketed_philly_share_of_reviews', 'bucketed_average_stars_given', 'bucketed_philly_reviews', 'bucketed_positive_reviews_over_total_reviews']
ohe_slim_philly_users = pd.get_dummies(slim_philly_users[slim_philly_users_columns])
ohe_slim_philly_users['user_id'] = slim_philly_users['user_id']
ohe_slim_philly_users
```

```python
# feather format to compress
ohe_slim_philly_users.to_feather('ohe_slim_philly_users.feather', compression='zstd')
```

```python
philly_bus['bucketed_positive_%'] = np.array(round(philly_bus['positive_%'], 1))
philly_bus['bucketed_positive_%'] = philly_bus['bucketed_positive_%'].astype(str)
philly_bus.head()
```

```python
bus_columns = ['bucketed_average_stars_received', 'bucketed_review_count', 'bucketed_positive_%', 'postal_code']
slim_philly_bus = philly_bus[bus_columns]
ohe_slim_philly_bus = pd.get_dummies(slim_philly_bus)
ohe_slim_philly_bus['business_id'] = philly_bus['business_id']
ohe_slim_philly_bus
```

```python
# feather format to compress
ohe_slim_philly_bus.to_feather('ohe_slim_philly_bus.feather', compression='zstd')
```
