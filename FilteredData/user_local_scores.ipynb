```python
philly_bus = pd.read_feather('business_philly.feather')
philly_reviews = pd.read_feather('review_philly.feather')
philly_users = pd.read_feather('user_philly.feather')
```

```python
from collections import Counter
```

```python
## Working on a function to return a dictionary of the cities each user reviewed in

user_cities = []
for sampled_user in philly_users.user_id:
    cities = []
    for biz_id in philly_reviews.business_id[philly_reviews.user_id == sampled_user]:
        cities.append(philly_bus.city[philly_bus.business_id == biz_id].values[0])
    user_cities.append((sampled_user, Counter(cities)))
user_cities
```

```python
philly_reviews_by_user = []
for i in user_cities:
    philly_reviews_by_user.append(i[1].most_common(1)[0][1])
len(philly_reviews_by_user)
```

```python
philly_users['philly_reviews'] = np.array(philly_reviews_by_user)
philly_users
```

```python
philly_users.review_count = philly_users.review_count.astype(int)
```

```python
philly_users['philly_share_of_reviews'] = philly_users.philly_reviews/philly_users.review_count
philly_users
```

```python
# feather format to compress
philly_users.to_feather('user_philly.feather')
```
